```python
data_encoded = pd.get_dummies(data, columns=['City', 'Profession', 'Sleep_Duration', 'Dietary_Habits', 'Degree'], drop_first=False)
data_encoded.head()
```

<h3><b>Dataset Splitting<u>(random Splitting)

```python
from sklearn.model_selection import train_test_split


X = data_encoded.drop('Depression', axis=1)  
y = data_encoded['Depression']
#70%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standardization
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data_encoded)
print("Standardized Data:\n", data_standardized)

# Normalization
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_encoded)
print("Normalized Data:\n", data_normalized)
```

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

model = LogisticRegression(max_iter=1000)  # max_iter is increased to ensure convergence

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy*100:.2f}%")

# Show detailed performance metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
#KNN 
from sklearn.neighbors import KNeighborsClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


ml_models = {
    "LogisticRegression": LogisticRegression(),
    " DecisionTreeClassifier": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "XGBoost": XGBRegressor(),
    "Naive Bayes": GaussianNB()
    
   
}
```

```python
with open('model_evaluation_output.txt', 'w') as f:

    
    for model_name, model in ml_models.items():
        f.write(f"\nModel Name: {model_name}\n")
        
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        
        if isinstance(model, (GaussianNB, DecisionTreeClassifier, KNeighborsClassifier,LogisticRegression)):  # If it's a classification model (like Naive Bayes) or  DecisionTreeClassifier
            y_pred = (y_pred > 0.5).astype(int)  # Convert probabilities to binary class labels
        
        # Write evaluation metrics for regression models
        if isinstance(model, ( DecisionTreeRegressor, RandomForestRegressor, GradientBoostingRegressor, XGBRegressor)):
            f.write(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred)}\n")
            f.write(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred)}\n")
            f.write(f"R-Squared Score (R2): {r2_score(y_test, y_pred)}\n")
        
        # Write evaluation metrics for classification models
        elif isinstance(model, (GaussianNB, DecisionTreeClassifier,KNeighborsClassifier,LogisticRegression)):
            f.write(f"Accuracy: {accuracy_score(y_test, y_pred)}\n")
            f.write(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}\n")
            f.write(f"Classification Report:\n{classification_report(y_test, y_pred)}\n")

print("Model evaluation saved to 'model_evaluation_output.txt'")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRegressor, XGBClassifier

# Define models
models = {
    "LogisticRegression": LogisticRegression(),
    " DecisionTreeClassifier": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "XGBoost": XGBRegressor(),
    "Naive Bayes": GaussianNB()
    
   
}

# Loop through each model, train and evaluate
for model_name, model in models.items():
    print(f'Model: {model_name}')
    
    model.fit(X_train, y_train)

    # Predictions
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    if isinstance(model, (DecisionTreeClassifier, KNeighborsClassifier, XGBClassifier, GaussianNB)):  # For classification models
        print(f'Training Accuracy: {accuracy_score(y_train, pred_train)}')
        print(f'Testing Accuracy: {accuracy_score(y_test, pred_test)}')
        print(f'Training Recall: {recall_score(y_train, pred_train)}')
        print(f'Testing Recall: {recall_score(y_test, pred_test)}')
        print(f'Training Precision: {precision_score(y_train, pred_train)}')
        print(f'Testing Precision: {precision_score(y_test, pred_test)}')
        print(f'Training F1score: {f1_score(y_train, pred_train)}')
        print(f'Testing F1score: {f1_score(y_test, pred_test)}')
    else:  # For regression models
        from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
        print(f'Training Mean Absolute Error: {mean_absolute_error(y_train, pred_train)}')
        print(f'Testing Mean Absolute Error: {mean_absolute_error(y_test, pred_test)}')
        print(f'Training Mean Squared Error: {mean_squared_error(y_train, pred_train)}')
        print(f'Testing Mean Squared Error: {mean_squared_error(y_test, pred_test)}')
        print(f'Training R2 Score: {r2_score(y_train, pred_train)}')
        print(f'Testing R2 Score: {r2_score(y_test, pred_test)}')

    # Cross-validation
    cross_validation_scores = cross_val_score(estimator=model, X=X_train, y=y_train, scoring='accuracy' if isinstance(model, (DecisionTreeClassifier, KNeighborsClassifier, XGBClassifier, GaussianNB)) else 'neg_mean_squared_error', n_jobs=-1, cv=5)
    print('Cross Validation Mean: ', cross_validation_scores.mean())
    
    # Confusion Matrix
    if isinstance(model, (DecisionTreeClassifier, KNeighborsClassifier, XGBClassifier, GaussianNB)):  # For classification models
        cm = confusion_matrix(y_test, pred_test)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
        disp.plot(cmap='YlGnBu', values_format='d')
        plt.title(f'Confusion Matrix: {model_name}')
        plt.show()
    
    print('-' * 60)
```
